```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Creating labeled examples
features = ['Support', 'Happiness', 'LogGDP', 'Freedom', 'Positive']
X = df[features]
us_life = df.loc[df['country'] == 'United States', 'Life'].values[0]


# Updated categories with one hot encoding, 1 is life expectancy greater than or equal to US, 
# 0 is less than the US
df['HighLifeExpectancy'] = (df['Life'] >= us_life).astype(int)
y = df['HighLifeExpectancy']
```

```python
# Inspect updated df
df.head()
```

```python
print("This is the shape of the dataframe: ", df.shape)
print("This is the shape of X: ", X.shape)
print("This is the shape of y: ", y.shape)
```

```python
# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.30, random_state=123)
```

```python
# Setting up possible hyperparameters

md = list(range(2, 10, 2))
msl = list(range(10,50,5))
ne = list(range(100, 500, 100))

param_grid = {'max_depth':md, 'min_samples_leaf':msl, 'n_estimators':ne}
```

```python
# Performing grid search to find possible parameters

print('Running Grid Search')

rf_class = RandomForestClassifier()
rf_grid = GridSearchCV(rf_class, param_grid, cv=3, scoring = 'neg_root_mean_squared_error', verbose = 1)
rf_grid_search = rf_grid.fit(X_train, y_train)

print('Finished Running Grid Search')
print('Updated')
```

```python
# Lower RMSE is better
rmse_RF = -1 * rf_grid_search.best_score_
print("RMSE for best model is : {:.2f}".format(rmse_RF))
```

```python
# Best parameters after grid 

rf_best_params = rf_grid_search.best_params_
rf_best_params
print(f"Optimal Max Depth: {rf_best_params['max_depth']}")
print(f"Optimal Min Samples Leaf: {rf_best_params['min_samples_leaf']}")
print(f"Optimal N Estimators: {rf_best_params['n_estimators']}")
```

```python
model = RandomForestClassifier(max_depth=rf_best_params['max_depth'], min_samples_leaf=rf_best_params['min_samples_leaf'], n_estimators=rf_best_params['n_estimators'])
model.fit(X_train, y_train)
class_label_predictions = model.predict(X_test)
acc_score = accuracy_score(y_test, class_label_predictions)
print("Accuracy:", round(acc_score, 4))
```

```python
conf_matrix = confusion_matrix(y_test, class_label_predictions)
print("Confusion Matrix After Classification (Test)")
print(f"True positive: {conf_matrix[0][0]} | False negative : {conf_matrix[0][1]}")
print(f"False positive: {conf_matrix[1][0]} | True negative : {conf_matrix[1][1]}")
```

```python
# Visualization of confusion matrix
plt.figure()
sns.heatmap(conf_matrix, annot=True, fmt='g', cmap='Purples', 
            xticklabels=['Predicted Low', 'Predicted High'], 
            yticklabels=['Actual Low', 'Actual High'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix Heatmap')
plt.show()
```

```python
# Plot the results/accuracy as a line graph

estimator_range = list(range(10, 201, 10))

# Removing outlier
estimator_range.pop(4)

accuracy_scores = []

for n_est in estimator_range:
    model = RandomForestClassifier(
        n_estimators=n_est,
        max_depth=rf_best_params['max_depth'], 
        min_samples_leaf=rf_best_params['min_samples_leaf'],
        random_state=123
    )
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    accuracy_scores.append(accuracy_score(y_test, predictions))

plt.figure()
plt.plot(estimator_range, accuracy_scores, marker='o', linestyle='-')
plt.xlabel('Number of Estimators (Trees)')
plt.ylabel('Accuracy Score')
plt.title('Random Forest Accuracy vs. Number of Trees')
plt.grid(True)
plt.show()
```
