```python
# YOUR CODE HERE
# defining features - simplifying their names
features = ['country', 'year', 'Life Ladder',

'Positive affect','Negative affect',
'Log GDP per capita', 'Social support',
'Healthy life expectancy at birth',
'Freedom to make life choices',
'Generosity', 'Perceptions of corruption']
df = df[features]

new_names_dict = {'Life Ladder': 'Happiness',
'Log GDP per capita': 'LogGDP',
'Social support': 'Support',
'Healthy life expectancy at birth': 'Life',
'Freedom to make life choices': 'Freedom',
'Perceptions of corruption': 'Corruption',
'Positive affect': 'Positive',
'Negative affect': 'Negative'}

df.rename(columns = new_names_dict, inplace = True)
df.head()
```

```python
# updated columns
list(df.columns)
```

```python
# remove missing features
nan_count = np.sum(df.isnull(), axis = 0)
print(f"Nan count before cleaning: {nan_count}")
df = df.dropna()
nan_count = np.sum(df.isnull(), axis = 0)
print(f"Nan count after cleaning: {nan_count}")
```

```python
# grouping by country
df = df.groupby(['country']).mean().reset_index()
df.head(15)
```

```python
# remove year feature
df = df.drop('year', axis=1)
```

```python
# updated cols
df.columns
```

```python
# scale data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df_to_scale = df.select_dtypes(float)
transformed_data = scaler.fit_transform(df_to_scale)
df_scaled = pd.DataFrame(transformed_data, columns = df_to_scale.columns, index = df_to_scale.index)
```

```python
df_scaled
```

```python
# hirearchical clustering
numerical_features = ['Happiness', 'LogGDP', 'Support', 'Life', 'Freedom','Generosity', 'Corruption', 'Positive', 'Negative']
sns.clustermap(df_scaled[numerical_features], method='average', metric='euclidean', figsize=(15,40))
# countries are clustered horizontally, countries with more similar features are closer together in distance
# features are clustered together vertically, more similar features are closer together
```

Based on the results of hierarchical clustering, we identified the 5 features most correlated to life expectancy. These features are the following:
1. Support
2. Happinesss
3. LogGDP
4. Freedom
5. Positive

In our graph above, features are clustered together vertically, such that features that are closer together are more strongly correlated. Based on this, we used the features closest to life expectancy on the graph. This initial exploration allowed us to select for the most relevant features to use for our classification model later on.

```python
# Creating Labeled examples
features = ['Support', 'Happiness', 'LogGDP', 'Freedom', 'Positive']
X = df[features]
y = df['Life']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.30, random_state=123)
```

```python
#Setting up possible hyperparameters

md = list(range(2, 10, 2))
msl = list(range(10,50,5))
ne = list(range(100, 500, 100))

param_grid = {'max_depth':md, 'min_samples_leaf':msl, 'n_estimators':ne}
```

```python
print('Running Grid Search')

rf_reg = RandomForestRegressor()
rf_grid = GridSearchCV(rf_reg, param_grid, cv=3, scoring = 'neg_root_mean_squared_error', verbose = 1)
rf_grid_search = rf_grid.fit(X_train, y_train)

print('Finished Running Grid Search')
print('Updated')
```

```python
rmse_RF = -1 * rf_grid_search.best_score_
print("RMSE for best model is : {:.2f}".format(rmse_RF))
```

```python
rf_best_params = rf_grid_search.best_params_
rf_best_params
```

```python
print('RF Implementation Starts')

rf_model = RandomForestRegressor(max_depth = 8, min_samples_leaf= 10, n_estimators = 200)
rf_model.fit(X_train, y_train)

print ('RF Done')
```
